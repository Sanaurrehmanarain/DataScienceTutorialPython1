# While loop

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
i = 1
n = 2
while i <= 10:
    print(f"{n} X {i} = {n * i}")
    i += 1
```

```python
l1 = [1, 2, 3, 4, 5] # Initializes a list with five elements
i = 0 # Sets an index variable to start from 0

while i < len(l1): # Loops as long as i is less than the length of l1 (i.e., 5)
    l1[i] = l1[i] + 100 # Updates the element at index i by adding 100
    i = i + 1 #  Increments i to move to the next index

l1
```

Step-by-step Execution:

- Initially: l1 = [1, 2, 3, 4, 5]
- Iteration 1 (i = 0): l1[0] = 1 + 100  [101, 2, 3, 4, 5]
- Iteration 2 (i = 1): l1[1] = 2 + 100  [101, 102, 3, 4, 5]
- Iteration 3 (i = 2): l1[2] = 3 + 100  [101, 102, 103, 4, 5]
- Iteration 4 (i = 3): l1[3] = 4 + 100  [101, 102, 103, 104, 5]
- Iteration 5 (i = 4): l1[4] = 5 + 100  [101, 102, 103, 104, 105]

After the loop, l1 = [101, 102, 103, 104, 105].

# For loop

Method 2 of showing image using markdown

![Alt text](<For loop.png>)

```python
l1 = ["mango", "apple", "banana", "grapes", "orange"]

for _ in l1:
    print(_)
```

```python
l2 = ["orange", "black", "white"]
l3 = ["chair", "table", "laptop"]

for i in l2:
    for j in l3:
        print(i, j)
```
