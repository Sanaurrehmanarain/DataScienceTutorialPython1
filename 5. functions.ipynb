```python
def num(x):
    if x % 2 == 0:
        print(f"{x} is even")
    else:
        print(f"{x} is odd")

num(2)
```

### Anonymous Functions

```python
# lambda
g = lambda x: x * x * x 
g(5)
```

```python
# filter
l1 = [87, 56, 90, 34, 2, 5, 7, 1, 8, 7, 98]

even_val = list(filter(lambda x: (x % 2 == 0), l1))
odd_val =  list(filter(lambda x: (x % 2 != 0), l1))

print(even_val)
print(odd_val)
```

```python
# map
l2 = [1, 2, 3, 4, 5, 6, 7, 8]

multiply = list(map(lambda x: x * 2, l2))
print(multiply)
```

```python
# reduce
from functools import reduce
sum = reduce(lambda x, y: x + y, l2)
sum
```
