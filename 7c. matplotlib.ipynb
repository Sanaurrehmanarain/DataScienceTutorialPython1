In `matplotlib.pyplot`, the function `plt.subplot(nrows, ncols, index)` is used to create a **subplot** grid in a figure. The parameters `(2, 1, 1)` and `(2, 1, 2)` define where each plot appears in that grid.

**Breakdown of** `plt.subplot(2, 1, 1)`:

   - `2`: total number of rows in the subplot grid

   - `1`: total number of columns in the subplot grid

   - `1`: index of the subplot (starts at 1, not 0)

So:

* `plt.subplot(2, 1, 1)` means **plot in the first position of a 2-row, 1-column grid**(i.e., the **top plot**).

* `plt.subplot(2, 1, 2)` means **plot in the second position of the 2-row, 1-column grid** (i.e., the **bottom plot**).

 Incorrect:

```python

plt.scatter(x, y, marker="*", color='blue', markersize=30)
```
`plt.scatter()` **does not use** `markersize`  that parameter is for `plt.plot()`.

 Correct:
```python

plt.scatter(x, y, marker="*", color='blue', s=300)  # Use 's' for marker size
plt.show()
```

 **Why?**
-   In `plt.scatter()`, the correct size parameter is:

    -   `**s**`  which stands for "size in points"

    -   So `s=300` is like a large star marker.

To show the order of points in a scatter plot (e.g., label each point with its sequence number), you can use plt.text() or plt.annotate() to add labels next to each point.

1. Using plt.text()

````python
# Scatter Plot
x = [4, 2, 5 , 7, 2, 4, 6, 2, 7, 9]
y = [9, 3, 5, 1, 8, 5, 4, 3, 7, 1]

plt.scatter(x, y, color='blue')

# Label each point with its sequence number using plt.text()
for i in range(len(x)):
    plt.text(x[i]+0.1, y[i]+0.1, str(i+1), fontsize=9, color='red')

plt.title("SScatter Plot with plt.text() Labels")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
````

 Explanation

-   `for i in range(len(x)):`  Loops through each point.

-   `plt.text(x[i]+0.1, y[i]+0.1, str(i+1), ...)`  Adds a label next to each point.

    -   `+0.1` offsets the label slightly so it doesnt overlap the point.

    -   `str(i+1)` converts the index into a label (starting from 1).

-   `fontsize, color`  Style the text.

2. Using plt.annotate()

````python
x = [4, 2, 5, 7, 2, 4, 6, 2, 7, 9]
y = [9, 3, 5, 1, 8, 5, 4, 3, 7, 1]

plt.scatter(x, y, color='green')

# Label each point with its sequence number using plt.annotate()
for i in range(len(x)):
    plt.annotate(
        str(i+1),                    # text
        (x[i], y[i]),                # point to label
        textcoords="offset points", # positioning method
        xytext=(5, 5),               # offset (x, y)
        ha='left',                   # horizontal alignment
        fontsize=9,
        color='purple'
    )

plt.title("Scatter Plot with plt.annotate() Labels")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
````

 Comparison

| Feature     | `plt.text()`                 | `plt.annotate()`                                   |
| ----------- | ---------------------------- | -------------------------------------------------- |
| Simplicity  |  Simple for basic labels   |  Slightly more code                               |
| Positioning | Manual offset (e.g., `+0.1`) | Uses `offset points` for relative placement        |
| Arrows      |  Not supported              |  Can add arrows                                  |
| Flexibility |  Less control               |  More advanced options (arrows, alignment, etc.) |


Below are some advanced `plt.annotate()` examples, including **arrows, custom alignment,** and **styling options**. These help highlight or explain points clearly in a scatter plot.

 Example: plt.annotate() with Arrows and Styling

````python
x = [4, 2, 5, 7, 2, 4, 6, 2, 7, 9]
y = [9, 3, 5, 1, 8, 5, 4, 3, 7, 1]

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='skyblue', edgecolors='black', s=100)

# Annotate each point with an arrow and a sequence number
for i in range(len(x)):
    plt.annotate(
        f"Point {i+1}",              # label text
        xy=(x[i], y[i]),             # the point to annotate
        xytext=(15, 10),             # offset label position (in points)
        textcoords="offset points",  # interpret xytext as offset in points
        ha='left',                   # horizontal alignment
        va='bottom',                 # vertical alignment
        fontsize=9,
        color='darkblue',
        arrowprops=dict(
            arrowstyle="->",         # arrow style (->, -[, <->, etc.)
            color='gray',
            lw=1.5,
            linestyle='--'
        )
    )

plt.title("Scatter Plot with Annotated Points and Arrows")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.tight_layout()
plt.show()
````

 **Key Options Explained**
-   `xy:` The data point to annotate.

-   `xytext:` Where to place the label (relative to the point).

-   `textcoords="offset points":` Interprets `xytext` as offset in points (rather than data units).

-   `arrowprops`: Dictionary to define the arrow style.

    -   `arrowstyle="->"`: Normal arrow

    -   `lw`: Line width

    -   `linestyle`: Solid, dashed (`--`), dotted (`:`), etc.

    -   `color`: Arrow color

 **Tips**
-   You can change `xytext` to control how far the label is from the point.

-   Try other `arrowstyle` values: "`->"`, `"-[", "|-|"`, `"fancy"`, `"wedge"`, etc.

-   You can also use `bbox` in `annotate()` to put a background behind the label.

 **Bonus: Fancy Label Box**

 Bonus: Fancy Label Box" refers to making the labels on your plot look more stylish and readable, by placing them inside shaped background boxes using the `bbox` parameter in `plt.annotate()`.

 **What is bbox in plt.annotate()?**

`bbox` stands for **bounding box**, and it's a dictionary that defines the style of a box around your annotation text. This can make labels stand out, especially when placed over busy plots.

````python
plt.annotate(
    f"Point {i+1}",
    xy=(x[i], y[i]),
    xytext=(10, 10),
    textcoords="offset points",
    ha='center',
    va='center',
    fontsize=8,
    color='black',
    bbox=dict(boxstyle="round,pad=0.3", fc="lightyellow", ec="black", lw=1),
    arrowprops=dict(arrowstyle="->", color="black", lw=1)
)
````

 Example: Fancy Label Box

````python

x = [2, 4, 6]
y = [5, 3, 8]

plt.scatter(x, y, color='lightcoral', s=100)

for i in range(len(x)):
    plt.annotate(
        f"Point {i+1}",
        xy=(x[i], y[i]),
        xytext=(10, 10),
        textcoords="offset points",
        ha='center',
        va='center',
        fontsize=9,
        color='black',
        bbox=dict(
            boxstyle="round,pad=0.3",  # Rounded box with padding
            fc="lightyellow",          # Fill color
            ec="black",                # Edge color
            lw=1                       # Line width
        ),
        arrowprops=dict(
            arrowstyle="->",
            color="black"
        )
    )

plt.title("Fancy Label Box Example")
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
# Sample data
fruit = ['Apple', 'Banana', 'Cherry', 'Date']
quantity = [30, 15, 45, 10]

# Outer pie (donut body)
plt.pie(quantity, labels=fruit, radius=2)

# Inner white circle (the hole)
plt.pie([1], radius=1, colors=['white'])

# Center title inside the donut
plt.text(0, 0, "Donut Chart", ha='center', va='center', fontsize=14, color='black')

plt.axis('equal')  # Keeps the donut circular
plt.show()
````

- `plt.text(0, 0, ...)`  Positions the text in the center of the chart.

- `ha='center', va='center'`  Horizontally and vertically aligns the text.

- `plt.axis('equal') ` Keeps the donut round, not oval (important for text alignment).

````python
# Sample data
fruit = ['Apple', 'Banana', 'Cherry', 'Date']
quantity = [30, 15, 45, 10]

# Calculate the total
total = sum(quantity)

# Outer pie (donut body)
plt.pie(quantity, labels=fruit, radius=2)

# Inner white circle (the hole)
plt.pie([1], radius=1, colors=['white'])

# Dynamic center text
plt.text(0, 0, f"Total:\n{total}", ha='center', va='center', fontsize=14, color='black')

plt.axis('equal')  # Keep circle aspect
plt.show()
````
