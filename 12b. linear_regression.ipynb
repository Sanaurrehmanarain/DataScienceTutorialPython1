```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
# Importing the main dataset
df = pd.read_csv("datasets/covid_19_india.csv", parse_dates=["Date"], dayfirst=True)
```

```python
df.head(3)
```

```python
# Keeping only required columns
df = df[["Date", "State/UnionTerritory", "Cured", "Deaths", "Confirmed" ]]

# Change the column names
df.columns = ["date", "state", "cured", "deaths", "confirmed"]
```

```python
df.tail(3)
```

```python
# Checking the new column names
for i in df.columns:
    print(i)
```

```python
today = df[df.date == "2020-06-22"]
today.head()
```

```python
#  Sorting data w.r.t number of confirmed death
confirmed_cases = today.sort_values(by="confirmed", ascending=False)
confirmed_cases.head()
```

```python
# Getting states with maximum number of confirmed cases
top_states_confirmed = confirmed_cases[0:5]
```

```python
top_states_confirmed
```

```python
# Making bar plot for states with top confirmed cases
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="state",
            y="confirmed",
            data=top_states_confirmed,
            hue="state")
plt.show()
```

```python
# Sorting data w.r.t number of death cases
death_cases = today.sort_values(by="deaths", ascending=False)
death_cases.head()
```

```python
top_states_death = death_cases[0:5]
```

```python
# Making bar plot for top states with deaths
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="state",
            y="deaths",
            data=top_states_death,
            hue="state")
plt.show()
```

```python
# Sorting data w.r.t number of cured cases
cured_cases = today.sort_values(by="cured", ascending=False)
cured_cases.head()
```

```python
top_states_cured = cured_cases[0:5]
```

```python
# Making bar plot for top states with cured cases
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="state",
            y="cured",
            data=top_states_cured,
            hue="state")
plt.show()
```

```python
# Linear Regression
from sklearn.model_selection import train_test_split
```

```python
# Converting date-time to ordinal
maha["date"] = maha["date"].map(dt.datetime.toordinal)
maha.head()
```

```python
# Getting dependent variable and independent variable
x = maha["date"]
y = maha["confirmed"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
y_train
```

```python
lr.fit (np.array (x_train).reshape (-1,1), np.array (y_train).reshape (-1, 1))
```
